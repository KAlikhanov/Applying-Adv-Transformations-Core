## Superhero_info.csv

### Splitting the Hero|Publisher column into two seperate columns.

```python
# Seeing how the column is formatted
super_info_df['Hero|Publisher'].head(3)
```

```python
# Splitting the column along the '|'
super_info_df[['Hero','Publisher']] = super_info_df['Hero|Publisher'].str.split('|',expand=True)
super_info_df.head(2)
```

```python
# Dropping the original column
super_info_df.drop(columns = ['Hero|Publisher'], inplace = True)
super_info_df.head()
```

### Converting the string column of dictionaries into dictionaries and splitting the dictionaries into seperate columns.

```python
# Converting the strings into dictionaries.
super_info_df['Measurements'] = super_info_df['Measurements'].str.replace("'",'"')
super_info_df['Measurements'] = super_info_df['Measurements'].apply(json.loads)
super_info_df['Measurements'].head()
```

```python
# Seperating the dictionary into two columns.
height_weight = super_info_df['Measurements'].apply(pd.Series)
height_weight
```

### Converting the strings to numeric values

```python
# Renaming the original Height and Weight columns for convenience later.
height_weight.rename(columns = {"Height": "H", "Weight": "W"}, inplace = True)
```

```python
# Splitting the height column along the ' '
height_weight[['Height','units']] = height_weight['H'].str.split(' ',expand=True)
height_weight.head(2)
```

```python
# Spliting the weight column along the ' '
height_weight[['Weight','units2']] = height_weight['W'].str.split(' ',expand=True)
height_weight.head(2)
```

```python
# Dropping the extra columns, the assignment doesnt say to keep units.
height_weight.drop(columns = ['H', 'W', 'units', 'units2'], inplace = True)
```

```python
# Converting our stripped columns into floats.
height_weight = height_weight.astype('float')
height_weight.info()
```

```python
height_weight.head()
```

### Concatonating the data back to the original dataframe.

```python
# Concatonating our new height_weight dataframe with the original one.
super_info_df = pd.concat([super_info_df, height_weight], axis = 1)
super_info_df.head()
```

```python
# Dropping the original column.
super_info_df.drop(columns = 'Measurements', inplace = True)
super_info_df.tail()
```

```python
super_info_df.info()
```

## Superhero_powers.csv

```python
super_power_df.tail()
```

### Turning the Powers column into a list

```python
super_power_df.loc[2, 'Powers']
```

```python
# Creating the column with the list
super_power_df['Powers_split'] = super_power_df['Powers'].str.split(",")
super_power_df.tail()
```

```python
# Exploding the list column
exploded = super_power_df.explode('Powers_split')
exploded[['hero_names', 'Powers', 'Powers_split']].tail(5)
```

```python
# Getting the columns to make for OHE
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

```python
# Looping and making the columns as well as the values
for col in cols_to_make:
    super_power_df[col] = super_power_df['Powers'].str.contains(col)
super_power_df.head()
```

```python
# Checking the new dataframe.
super_power_df.tail()
```

### Merging the two dataframes into one final dataframe

```python
# Merging the dataframes with an outer join.
super_merged_df = pd.merge(super_power_df, super_info_df, left_on = 'hero_names', right_on = 'Hero', how = 'outer')
```

```python
super_merged_df.head()
```
